```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, y, 'b')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y, 'r--')
axes[1].plot(y, x, 'g--')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2) # unpacking the axes array into separate variables and naming them ax0 and ax1
ax0.plot(x, y, 'r--')
ax1.plot(y, x, 'g--')
```
