```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 5, 11)
y = x**2
```

```python
plt.plot(x, y, 'rs-')
plt.show()
```

```python
plt.hist(x, y,)
plt.show()
```

```python
plt.scatter(x, y,)
plt.show()
```

```python
plt.bar(x, y,)
plt.show()
```

```python
plt.pie(x)
plt.show()
```

```python
plt.boxplot( y)
plt.show()
```
