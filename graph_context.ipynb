```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(10, 5)) # unpacking the axes array into separate variables and naming them ax0 and ax1
ax0.plot(x, y, 'r', label='Sine Wave')
ax0.set_title('Sine Function')
ax0.set_xlabel('X-axis')
ax0.set_ylabel('Y-axis')
ax0.legend()
ax1.plot(y, x, label='Inverse Sine', color='green', linestyle='--')
ax1.set_title('Inverse Sine Function')
ax1.set_xlabel('Y-axis')
ax1.set_ylabel('X-axis')
ax1.legend()
fig.tight_layout() # Adjusts the spacing between subplots to prevent overlap
```

```python
plt.plot(x, y, label=('Sine Wave'))
plt.title('Sine Function')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(loc='upper center')
```
